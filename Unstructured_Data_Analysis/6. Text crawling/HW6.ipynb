### text crawling

- 우리학교 유튜브에서 2021학년도 1학기와 2학기 DS Interface에 올라온 논문들의 논문제목을 크롤링해서 아래의 예시와 같이 출력해주시면 됩니다 
- 동영상 앞에 [DS Interface]라고 표기된 부분은 제외하고 제목만 출력

```python
driver1 = wd.Chrome(r'C:\Users\정주현\Downloads/chromedriver.exe')
driver1.get(url1)
driver1.implicitly_wait(3)

soup = BeautifulSoup(driver1.page_source, 'html.parser')
driver1.close()
```

```python
term1 = soup.body.find_all('a', id='video-title')
```

```python
term1_list = []
for title in term1:
    term1_list.append(title.text.strip().strip('[DS Interface] '))
```

```python
pd.DataFrame(term1_list, columns=['2021학년도 1학기'])
```
