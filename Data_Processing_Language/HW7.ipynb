```python
import pandas as pd
import numpy as np
import time
import re
from selenium import webdriver as wd
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
```

```python
# 검색어 입력받기
search_word = input("검색어를 입력해 주세요 : ")
```

```python
url = 'https://patents.google.com/'
```

### text crawling

```python
driver = wd.Chrome(r'C:\Users\정주현\OneDrive - 서울과학기술대학교\바탕 화면/chromedriver.exe')
driver.implicitly_wait(3)

soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
driver.get(url) # 크롤링하고자 하는 사이트 주소 입력

search = driver.find_element_by_xpath('//*[@id="searchInput"]') # 검색 창 xpath 
search.clear()
search.send_keys(search_word) # 입력 받은 검색어 검색창에 입력
time.sleep(5)
search.send_keys(Keys.ENTER) # 엔터 키 누름
```

```python
# 25선택
driver.find_element_by_xpath('//*[@id="count"]/div[2]/span[2]/span[4]/dropdown-menu/span/iron-icon[1]').click()
time.sleep(3)
driver.find_element_by_xpath('/html/body/search-app/search-results/search-ui/div/div/div[2]/div/div/div[1]/div[2]/div[2]/span[2]/span[4]/dropdown-menu/iron-dropdown/div/div/div/div[2]').click()
```

```python
title = []
country = []
patent = []
inventor = []
for _ in range(25):
    driver.find_element_by_xpath(f'/html/body/search-app/search-results/search-ui/div/div/div/div/div/div[1]/section/search-result-item[{_+1}]/article/state-modifier/a/h3/raw-html/span').click()
    time.sleep(2)
    title.append(driver.find_element_by_xpath('/html/body/search-app/search-result/search-ui/div/div/div/div/div/result-container/patent-result/div/div/div/h1').text)
    country.append(driver.find_element_by_xpath('//*[@id="wrapper"]/div[1]/div[2]/section/header/p').text)    
    patent.append(driver.find_element_by_xpath('//*[@id="pubnum"]').text)
    inventor.append(driver.find_element_by_xpath('/html/body/search-app/search-result/search-ui/div/div/div/div/div/result-container/patent-result/div/div/div/div[1]/div[2]/section/dl[1]/dd[1]/state-modifier/a').text)
    
    
    back = driver.find_element_by_xpath('//*[@id="compactQuery"]/span[1]/iron-icon')
    back.click()
    
```

```python
pd.DataFrame({'제목':title,'국가':country,'특허번호':patent,'inventor':inventor})
```
